```python
import pandas as pd
import numpy as np
```

### Process the Toronto Neighbourhood profiles data

```python
tpdf=pd.read_csv('neighbourhood-profiles-csv.csv')
tpdf=tpdf.T
tpdf.drop(tpdf.index[[0,1,2,3,5]], inplace=True)
tpdf.iloc[0]=tpdf.iloc[0].str.strip()
tpdf.rename(index={'Characteristic':'Neighbourhood'},inplace=True)
tpdf.columns = tpdf.iloc[0]
tpdf.drop(tpdf.index[[0]], inplace=True)
tpdf.insert(0, "Index", np.arange(140), True)
neighbourhood=tpdf.index
tpdf.set_index('Index', drop=True,inplace=True)
tpdf.insert(0, "Neighbourhood", neighbourhood, True)
df=tpdf[["Neighbourhood","Population density per square kilometre","Divorced","Without employment income","Without after-tax income","Never married","Separated","Widowed","5 or more persons","Immigrants"]]
status=pd.to_numeric(df['Divorced'])+pd.to_numeric(df['Separated'])+pd.to_numeric(df['Never married'])+pd.to_numeric(df['Widowed'])
df.insert(2, "Status", status, True)
df=df.drop(["Divorced","Widowed","Separated","Never married"],axis=1)
df = df.loc[:,~df.columns.duplicated()]
```

### Process the Toronto Covid data

```python
covid=pd.read_csv('COVID19 cases.csv')
```

```python
covid_summary=covid.groupby("Neighbourhood Name").size().reset_index(name='Count')
```

```python
covid_summary = covid_summary.rename(columns={'Neighbourhood Name': 'Neighbourhood'})
```

### Process the Toronto Assault data

```python
assaults=pd.read_csv('neighbourhood-crime-rates.csv')
```

```python
assaults=assaults[['Neighbourhood','Assault_2020']]
```

```python
assaults = assaults.rename(columns={'Assault_2020':'Assault'})
```

### Toronto Neighbourhood profiles data with Covid Data

```python
result=pd.merge(left=df, right=assaults, left_on='Neighbourhood', right_on='Neighbourhood')
result=pd.merge(left=result, right=covid_summary, left_on='Neighbourhood', right_on='Neighbourhood')
```

```python
result['Population density per square kilometre'] = result['Population density per square kilometre'].str.replace(",","").astype(int)
result['Without employment income'] = result['Without employment income'].str.replace(",","").astype(int)
result['Without after-tax income'] = result['Without after-tax income'].str.replace(",","").astype(int)
result['Immigrants'] = result['Immigrants'].str.replace(",","").astype(int)
result['5 or more persons'] = result['5 or more persons'].str.replace(",","").astype(int)
```

### Drop Population density per square kilometre because it is not correlated with the Data

```python
result.drop(['Population density per square kilometre'],axis=1,inplace=True)
```

### Create Groups (Low=Low Covid infection, Medium: Medium COVID infection, High COVID infection)

```python
result.insert(8, "Group", '', True)
```

```python
result.loc[result['Count'] <= 500, ['Group']] = 'Low'
```

```python
result.loc[(result['Count'] > 500) & (result['Count'] <= 1000),['Group']]='Medium'
```

```python
result.loc[result['Count'] > 1000, ['Group']] = 'High'
```

```python
import geopandas
gp = geopandas.read_file('Neighbourhoods.zip')
```

```python
geo_data=covid[['Neighbourhood Name','FSA']]
geo_data=geo_data.dropna(axis='rows')
geo_data
geo_data=geo_data.drop_duplicates()
geo_data=geo_data.rename(columns={'Neighbourhood Name': 'Neighbourhood'})
```

```python
postal_codes=pd.read_csv('postalcode.csv')
postal_codes=postal_codes.drop(['Borough','Neighborhood'],axis=1)
```

```python
geo_data=pd.merge(left=geo_data, right=postal_codes, left_on='FSA', right_on='PostalCode')
```

```python
geo_data=pd.merge(left=geo_data, right=result, left_on='Neighbourhood', right_on='Neighbourhood')
```

```python
geo_data=geo_data[['Neighbourhood','Latitude','Longitude','Group']]
```

```python
geo_data=geo_data.drop_duplicates()
```

```python
geo_data = geopandas.GeoDataFrame(
    geo_data, geometry=geopandas.points_from_xy(geo_data.Longitude, geo_data.Latitude))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
%matplotlib inline

fig, ax = plt.subplots(figsize=(14,12))

gp.plot(ax=ax,facecolor='white',edgecolor='lightgray')

group0=geo_data[geo_data.Group=='Low']
group0.plot(ax=ax, color='green')

group1=geo_data[geo_data.Group=='Medium']
group1.plot(ax=ax, color='yellow')

group2=geo_data[geo_data.Group=='High']
group2.plot(ax=ax, color='red')
```
