```python
import pandas as pd
import numpy as np
```

### Toronto Neighbourhood profiles data (That has data about neighberhoods unemployment,low-income,Youth rate etc..)

```python
tpdf=pd.read_csv('neighbourhood-profiles-csv.csv')
tpdf=tpdf.T
tpdf.drop(tpdf.index[[0,1,2,3,5]], inplace=True)
tpdf.iloc[0]=tpdf.iloc[0].str.strip()
tpdf.rename(index={'Characteristic':'Neighbourhood'},inplace=True)
tpdf.columns = tpdf.iloc[0]
tpdf.drop(tpdf.index[[0]], inplace=True)
tpdf.insert(0, "Index", np.arange(140), True)
neighbourhood=tpdf.index
tpdf.set_index('Index', drop=True,inplace=True)
tpdf.insert(0, "Neighbourhood", neighbourhood, True)
df=tpdf[["Neighbourhood","Population density per square kilometre","Youth (15-24 years)","Without employment income","Without after-tax income","5 or more persons","Immigrants","Total visible minority population"]]
df = df.loc[:,~df.columns.duplicated()]
```

### Toronto crime data

```python
assaults=pd.read_csv('neighbourhood-crime-rates.csv')
```

```python
assaults=assaults[['Neighbourhood','Assault_2020']]
```

```python
assaults = assaults.rename(columns={'Assault_2020':'Assault'})
```

## Merge profile data & crime data

```python
result=pd.merge(left=df, right=assaults, left_on='Neighbourhood', right_on='Neighbourhood')
```

```python
result['Population density per square kilometre'] = result['Population density per square kilometre'].str.replace(",","").astype(int)
result['Without employment income'] = result['Without employment income'].str.replace(",","").astype(int)
result['Without after-tax income'] = result['Without after-tax income'].str.replace(",","").astype(int)
result['Immigrants'] = result['Immigrants'].str.replace(",","").astype(int)
result['Youth (15-24 years)'] = result['Youth (15-24 years)'].str.replace(",","").astype(int)
result['Total visible minority population'] = result['Total visible minority population'].str.replace(",","").astype(int)
```

```python
result.head(5)
```

```python
result.to_csv('toronto-youth-data.csv')
```
